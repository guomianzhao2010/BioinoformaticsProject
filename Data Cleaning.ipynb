```python
import pandas as pd
```

```python
#import csv files using pandas
demographics=pd.read_csv("Demographics.csv")
labs=pd.read_csv("labsfirstday_ARDS.csv")
vitals=pd.read_csv("Vitals_mimic.csv")
```

Data Cleaning: Demographics 

```python
#Take a look at the demographics dataframe
demographics.head()
```

```python
#Remove all the patients that had multiple admissions. After removal, we still have 6749 unique patients 
demographics=demographics.drop_duplicates(subset=['subject_id'], keep=False)

#Drop a few columns that we don't need
demographics=demographics.drop(columns=['row_id','has_chartevents_data'])

# Rename the death column. 1=death, 0=alive
demographics.rename(columns={'hospital_expire_flag':'death'}, inplace=True)
```

Data Cleaning: Labs

```python
#Take a look at the dataframe
labs.head()
```

```python
#Only keep lab values of patients we have in the demographics dataframe
labs=labs[labs.subject_id.isin(demographics.subject_id)]
```

Data Cleaning: Vitals

```python
vitals.head()
```

```python
#Only keep lab values of patients we have in the demographics dataframe
vitals=vitals[vitals.subject_id.isin(demographics.subject_id)]
```
